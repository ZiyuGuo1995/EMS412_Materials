## Double Integrals and Area Between Curves in Python

By using **dblquad()** method, we can get the double integration of a given function from limit $a$ to $b$ by using scipy.


Integrate. dblquad() method. 


> Return : Return the double integrated value of a polynomial


Before doing it, we want to know the lambda function. "lambda" function is a small anonymous function which returns an object.

The object returned by lambda is usually assigned to a variable or used as a part of other bigger functions.

Instead of the conventional def keyword used for creating functions, a lambda function is defined by using the lambda keyword. The structure of lambda can be seen below:

```python
# A squaring lambda function x^2
square = lambda x : x**2
num = square(5)
print(num)
```

## 01 - Integrate:

$   f(x,y)=\int_{0}^{2}\int_{0}^{x^2}\ y\ dydx  $


```python
from scipy.integrate import dblquad
import numpy as np

def integrand(y, x):
    'y must be the first argument, and x the second.'
    return y

ans, err = dblquad(integrand, 0, 2,        # 0 and 2 is the lower and upper bound
                   lambda x: 0,            # lower bound for the y is 0
                   lambda x: x**2)         # lower bound 
print('ans=', ans)
print('err=', err)
```

 ## 02 - Integrate:

$   f(x,y)=\int_{0}^{1}\int_{0}^{x^2}\ (x+2)\ dydx  $

```python
from scipy.integrate import dblquad
import numpy as np

def integrand(y, x):
    'y must be the first argument, and x the second.'
    return  x+2

ans, err = dblquad(integrand, 0, 1,
                   lambda x: 0,
                   lambda x: x**2)
print('ans=', ans)
print('err=', err)
```

```python
from scipy.integrate import dblquad
import numpy as np

def integrand(y, x):
    'y must be the first argument, and x the second.'
    return  1

ans, err = dblquad(integrand, 2, 4,
                   lambda x: 2,
                   lambda x: 6)
print( ans)
```

 ## 04 - Integrate:

$   f(x,y)=\int_{0}^{2}\int_{x}^{6-x^2}\ (3x+2y)\ dydx  $

```python
from scipy.integrate import dblquad
import numpy as np

def integrand(y, x):
    'y must be the first argument, and x the second.'
    return  (3*x+2*y)

ans, err = dblquad(integrand, 0, 2,
                   lambda x: x,
                   lambda x: 6-x**2)
print( ans)
```

 ## 05 - Determine the area in the First quadrant of the region bounded by :

### Given parable: $ f(x)=x^3 $ 

### And Line: $ g(x)=4x $ 

$ over $ 

$ (0<=x<=2) $


```python
x = np.linspace(0, 3, 1000)
def f(x): return x**3
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(x, f(x))
plt.axhline(color = 'black')
plt.fill_between(x, f(x), where = [(x > 0) and (x < 2) for x in x], color = 'green', alpha = 0.3)
```

```python
import sympy as sy
x = sy.Symbol('x')
```

```python
def g(x): return 4*x
```

```python
x = np.linspace(0, 3, 100)
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.fill_between(x, f(x), g(x),  where = [(x >= 0) and (x <= 2) for x in x], color = 'green', alpha = 0.2)
```
