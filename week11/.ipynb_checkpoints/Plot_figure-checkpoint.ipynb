```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
######################################################
dir = './'
######################################################
    
data = pd.read_csv(dir+'data.csv')
    
######################################################
####The header of each columar is associate with the data. For example, "data.LE11" means it will call the column for the LE11
data.LE11 = data.LE11.astype(float)
data.S11 = data.S11.astype(float)
data.LE22 = data.LE22.astype(float)
data.S22 = data.S22.astype(float)
data.LE12 = data.LE12.astype(float)
data.S12 = data.S12.astype(float)
```

```python
# LE11, S11, LE22, S22, LE12, S12 = np.loadtxt('stress-strain.txt', delimiter=',', usecols=(0, 6), unpack=True,  skiprows=1)
```

```python
fig, ax = plt.subplots()
ax.plot(data.LE11, data.S11)

ax.set(xlabel= r"Strain, $\epsilon_{11}$", ylabel=r"Stress, $\sigma_{11}$ (MPa)",
       title='Stress-strain curves')
#ax.grid()

fig.savefig("S11.pdf")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(data.LE22, data.S22)

ax.set(xlabel= r"Strain, $\epsilon_{22}$", ylabel=r"Stress, $\sigma_{22}$ (MPa)",
       title='Stress-strain curves')
plt.xlim(-0.04,0)
#ax.grid()

fig.savefig("S22.pdf")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(data.LE12, data.S12)

ax.set(xlabel= r"Strain, $\epsilon_{12}$", ylabel=r"Stress, $\tau_{12}$ (MPa)",
       title='Stress-strain curves')
#ax.grid()

fig.savefig("S12.pdf")
plt.show()
```

```python
######################################################
dir = './'
######################################################
    
data2 = pd.read_csv(dir+'Tension_45.csv')
    
######################################################
####The header of each columar is associate with the data. For example, "data.LE11" means it will call the column for the LE11
data2.RF2 = data2.RF2.astype(float)
data2.U2 = data2.U2.astype(float)
```

```python
fig, ax = plt.subplots()
ax.plot(data2.U2, data2.RF2)

ax.set(xlabel= r"Displacement, $\delta_{2}$", ylabel=r"Load, $P_2$ (N)",
       title='Stress-strain curves')
#ax.grid()

fig.savefig("S12.pdf")
plt.show()
```
