## 1. Plot a stress-strain curve from a given function

Let's start from a simplest case: linearly-elastic materials, for example ceramics. 

For a given ceramics materials, we found its Young's modulus $E=400$ GPa. Let's plot its stress-strain curve. 

We know the stress-strain relationship from Hooke's law: $\sigma=E\epsilon$, here we know the modulus already. So this line function can be plot directly. 

The first thing is to generate an array of strain values using the **numpy.linspace**

Introduction: numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)

Return evenly spaced numbers over a specified interval.

Returns num evenly spaced samples, calculated over the interval

The endpoint of the interval can optionally be excluded.

### 1.1 Generate strain values from 0 to 0.02, with 100 interval

```python
strain=np.linspace(0, 0.02, num=100, endpoint=True)
print(strain) # print it out to check the values, you can add a # in front to comment it out
```

### 1.2 Calculate the stress

```python
E=400*1e3 # MPa, we normally use MPa in the calculation
stress=E*strain # MPa, stress normally use MPa, as it is convient to show the magnitude. 1 MPa = 10^6 Pa
# print(stress)
```

### 1.3 Plot the figure

```python
fig, ax = plt.subplots() # Create a figure and a set of subplots, by default, it will be one figure
# plt.subplots(2, 2) means generation of 2 columns and 2 rows of figures
ax.plot(strain, stress)  # Strain is the x-axis data and stress is y-axis data
ax.set_xlim(0, 0.03)  # Set the limit of x axis
ax.set_ylim(0, 1e4)  # Set the limit of y axis

# Add axis labels, title
ax.set(xlabel= r"strain, $\epsilon$", ylabel=r"Stress, $\sigma$ (MPa)",
       title='Stress-strain curve')

# Switch grid on or off
#ax.grid() 

# fig.savefig("Figure1.pdf") # Comment it out if you want to save the figure as a PDF file
plt.show()
```

## 2. Plot a stress-strain curve from the measured data

Suppose we have done a tensile test on a dogbone sample, the key paramters we will use are the gauage length, gauge width, sample thickness. See the figure below:



After the test, we have obtained some force and displacement data and save it in a file called "data.csv". 

Now we need to plot stress-strain curve. See the example below to plot stress-strain curve

## 2.1 Read the data file

```python
######################################################
dir = './' # This define the current directory is in the same folder of the ipynb file
######################################################
    
data = pd.read_csv(dir+'data.csv')  ## data.csv files 
    
######################################################
####The header of each columar is associate with the data. For example, "data.LE11" means it will call the column for the LE11
Force = data.RF2.astype(float) # Read the data under the name "RF2"
Displacement = data.U2.astype(float) # Read the data under the name "U2"
```

Now we need to convert the load and displacement to stress and strain. 

Given the gauge length $l_0=50 $ mm, gauge width $w_0=13 $ mm and sample thickness $t=3.2 $ mm, now we can use them to calculate stress and strain. 

Remember: stress= Force/cross-section area  

$\sigma= \frac{F}{A}$

and strain = change of sample length / orginal length

$\epsilon = \frac{\Delta l}{l_0}$

## 2.2 Define the parameters

```python
l0 = 50 #unit: mm
w0 = 13 #unit: mm
t = 3.2 #unit:mm
# Convert force to stress
A = w0*t # cross-section area
stress = Force / A  # Unit: N/mm^2, which is equivalent to MPa

# Convert displacement to strain
strain = Displacement / l0 # Unit: Dimensionalness 
```

## 2.3 Plot the figures

```python
fig, ax = plt.subplots()
ax.plot(strain, stress)  # Strain is the x-axis data and stress is y-axis data
ax.set_xlim(0, 0.2)  # Set the limit of x axis
ax.set_ylim(0, 120)  # Set the limit of y axis

# Add axis labels, title
ax.set(xlabel= r"strain, $\epsilon$", ylabel=r"Stress, $\sigma$ (MPa)",
       title='Stress-strain curve')

# Switch grid on or off
#ax.grid() 

# fig.savefig("Figure2.pdf")
plt.show()
```
