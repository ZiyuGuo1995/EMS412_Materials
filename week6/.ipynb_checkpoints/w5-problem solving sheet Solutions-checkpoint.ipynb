## Q1: Construct a force vector

Determine the components along the x and y axis of the vector F, and construct a 2D force vector using Numpy.array.



```python
## Your Code ## 
def force_vec(F, theta): # theta from +ve x axis
    Fx = F*cos(theta*pi/180)
    Fy = F*sin(theta*pi/180)
    V = np.array([Fx, Fy])
    return V
```
