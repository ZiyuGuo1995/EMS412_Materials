# 1. Plot the internal shear force and moment diagram

In this week, the learning outcome is to plot the internal shear force and moment diagram using Python.
Most of the calculations can be done by hand. 

### Example 1

Given a simpley supported beam, plot its internal shear force and moment diagrams along the x-axis. The beam has a length of $L=$ 200 mm. The applied concentrated load is $P=$ 1000 N. 

### 1. Define the paramters

```python
## Your code ## 
## Find the example in the tutorial sheet

import numpy as np
import matplotlib.pyplot as plt

#Beam parameters
L=200 #mm, length
a = 100 #mm, to change the position of the point load P
b = L - a # mm

h=100 #mm, height
P=1000 #N, point load P
```

### 2. Find the support reaction forces

The support reaction force:

$F^B_y=P a/L$ 

$F^A_y=-P-F^B_y$ 

```python
## Reaction forces:
FB_y = -(P*a)/L
FA_y = -P - FB_y
```

### 3. Do the Cut 1 $(0 \leq x \leq a)$ and find the expression

The internal shear force in terms of $x$:

$V(x)=-F^A_y$ 

The moment in terms of $x$:

$M(x)=V(x)\ x$ 

```python
x1=np.linspace(0,a, num=100) # define the range from 0 to distance of a
V1 = -FA_y # This is for V(x)
M1 = V1*x1 # This is for M(x)

#Max Values
maxV = abs(V1)
maxM = abs(V1*a)
```

### 4. Do the Cut 2 $(a \leq x \leq l)$ and find the expression

The internal shear force in terms of $x$:

$V(x)=-P-F^A_y$ 

The moment in terms of $x$:

$M(x)= Pa+V(x)\ x$ 

```python
x2=np.linspace(a,L, num=100) # Define the range for x from a to L
V2 = -P-FA_y # This is for V(x)
M2 = P*a+V2*x2 # This is for M(x)
```

### 5. Plot the diagram 

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 6))

# The first diagram
ax1.fill_between(x1, 0, V1, facecolor='#377EB8', alpha=0.8)
ax1.fill_between(x2, 0, V2, facecolor='#377EB8', alpha=0.8)
ax1.set_ylabel('Shear force (N)')
ax1.text(0.78, 0.9, textstr1, transform=ax1.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)
ax1.set_title('Cantilever Beam Subjected to Gravity Load')
ax1.grid (True)

# The second diagram
ax2.fill_between(x1, 0, M1, facecolor='#55BA87', alpha=0.8)
ax2.fill_between(x2, 0, M2, facecolor='#55BA87', alpha=0.8)
ax2.set_ylabel('Moment (N*mm)')
ax2.text(0.68, 0.68, textstr2, transform=ax2.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)
ax2.grid (True)

textstr1 = 'max=%.2f (N)'%(maxV) # Show the maximum value for shear force
textstr2 = 'max=%.2f (N*mm)'%(maxM) # Show the maximum value for the moment
props = dict(boxstyle='round', facecolor='white')


plt.show()
#plt.savefig('beam.pdf', format='svg', dpi=1200) # Export the figure
```
