## Q1: Exercise on the mid-point rule

Compute by hand the area composed of two rectangles (of equal width) that approximates the integral $f(x)$ below. Make a test function that calls the midpoint function in midpoint.py and compares the return value with the hand-calculated value.

$f(x)=\int _1^3 2x^3dx$

### Your code below:

#### Your computed value by hand

$f(x)=\int _1^3 2x^3dx=\frac{2}{4}x^4|^3_1=\frac{1}{2}x^4|^3_1=\frac{1}{2}(3^4-1^4)=40$

```python
## Your own Integration function, check the tutorial to see how you can do it
def Integrate (N, a, b):
    def f(x):
        # type your function after return
        f= 2*x**3   ## This is area you define the function 
        return f # Here this is a linear function, try square function later, you can put your function here
    value=0  # initial value
    TotalArea=0 # initial value
    for n in range(1, N+1): # Here range to N+1, not N
        x_bar = a+(2*n-1)*(b-a)/(2*N)
        value = value + f(x_bar) # sum up all thea values
    TotalArea= ((b-a)/N)*value # Approxiamted area
    return TotalArea    
```

```python
## Test the function
N=10
a=1
b=3
print ("Here is your answer using sinine function")
print (Integrate (N, a, b))
```

```python
## Error analysis
# Absolute error
Correct_answer=40
Midpoint_prediction=Integrate (N, a, b)

Ab_error=abs(Midpoint_prediction-Correct_answer)

# Relative error
Relative_error=abs(Midpoint_prediction-Correct_answer)/Correct_answer*100 # Percentage

print('Absolute error is', Ab_error, 'using iteration number of ', N)
print('Relative error is', Relative_error,'%', 'using iteration number of ', N)
```

## Q2: Implement of mid-point integration


Find the definite integral of 

$$\int_0^2 {\sin (x){e^{{x^2}\cos (x)}}dx} $$

using the mid-point rule.

```python
## Your own Integration function, check the tutorial to see how you can do it
def Integrate (N, a, b):
    def f(x):
        # type your function after return
        f= sin(x)*exp(x**2*cos(x))   ## This is area you define the function 
        return f # Here this is a linear function, try square function later, you can put your function here
    value=0  # initial value
    TotalArea=0 # initial value
    for n in range(1, N+1): # Here range to N+1, not N
        x_bar = a+(2*n-1)*(b-a)/(2*N)
        value = value + f(x_bar) # sum up all thea values
    TotalArea= ((b-a)/N)*value # Approxiamted area
    return TotalArea    
```

```python
## Test the function
N=200
a=1
b=3
print ("Here is your answer using sinine function")
print (Integrate (N, a, b))
```

```python
## Error analysis
# Absolute error
Correct_answer=40
Midpoint_prediction=Integrate (N, a, b)

Ab_error=abs(Midpoint_prediction-Correct_answer)

# Relative error
Relative_error=abs(Midpoint_prediction-Correct_answer)/Correct_answer*100 # Percentage

print('Absolute error is', Ab_error, 'using iteration number of ', N)
print('Relative error is', Relative_error,'%', 'using iteration number of ', N)
```

## Q3: Numerical integration using the Trapezoidal Rule

We can also approximate the value of a definite integral by using trapezoids rather than rectangles. In Figure below , the area beneath the curve is approximated by trapezoids rather than by rectangles.

The trapezoidal rule for estimating definite integrals uses trapezoids rather than rectangles to approximate the area under a curve. To gain insight into the final form of the rule, consider the trapezoids shown in Figure above. We assume that the length of each subinterval is given by  x . First, recall that the area of a trapezoid with a height of  h  and bases of length  b1  and  b2  is given by  Area=0.5*h(b1+b2) . We see that the first trapezoid has a height  x  and parallel bases of length  f(x0)  and  f(x1) . Thus, the area of the first trapezoid in Figure  2.5.2  is

Let ${\displaystyle \{x_{n}\}}$ be a partition of ${\displaystyle [a,b]}$ such that ${\displaystyle a=x_{0}<x_{1}<\cdots <x_{n-1}<x_{n}=b}$ and ${\displaystyle \Delta x_{n}}$
be the length of the n-th subinterval (that is, ${\displaystyle \Delta x_{n}=x_{n}-x_{n-1}}$), then

${\displaystyle \int _{a}^{b}f(x)\,dx\approx \sum _{n=1}^{N}{\frac {f(x_{n-1})+f(x_{n})}{2}}\Delta x_{n}={\tfrac {\Delta x}{2}}\left(f(x_{0})+2f(x_{1})+2f(x_{2})+2f(x_{3})+2f(x_{4})+\cdots +2f(x_{n-1})+f(x_{n})\right).}$

## Solution

```python
def InterageTrapz(N, a, b):
    def f(x):
        ## Type your function after the return
        return 2*x**3 
    value=0
    TotalArea=0
    dx=(b-a)/N # delta x, the interval length
    # value = value + a + b
    value = value + f(a) + f(b) # f(x_0)=a, f(x_n)=b, add the first and last value
    for n in range (1, N): # This will iterate number from x_1 to x_(N-1), as the last value has been caculated
        xn=a+n*dx # This give the current x_n value
        value += 2*f(xn)
    TotalArea = value*dx/2 # This should be outside the loop
    return TotalArea 
```

```python
N=10
a=1
b=3
print ("Here is your answer using Trapzoidal rule")
print (InterageTrapz (N, a, b))
```

Our basic integration program has the inconvenience of depending on the number of trapeziums that we have to change manually. As a basic rule, if we double the number of trapeziums and get the same answer within 0.01, the answer is probably correct.

## Q4: Run the program. How many trapeziums are needed to get a relative error of less than 0.01 (or 1e-02)?
We can do this as follows:

## Solution

```python
from math import *
N = 20
# Because we don't know the number N, we use while to find it iteratively.
# "While" is very similar with "for" loops, but don't need a range
while abs(InterageTrapz(N, a,b) - InterageTrapz(N*2, a,b)) > 0.01:
    N += 1
print(N)
```

# Q5. Use numerical integration to calculate the second moment of area

The second moment of area (also known as area moment of inertia) is a geometrical property of an area which reflects how its points are distributed with regard to an arbitrary axis. 

Consider a rectangle with base b and height h whose centroid is located at the origin. $I_{x}$ represents the second moment of area with respect to the x-axis; $I_{y}$ represents the second moment of area with respect to the y-axis; Here $h=3$ mm and $b=9$ mm.Now calculate the 2nd moment of area $I_{x}$.



The key point here is to determine the area of a small elment $dA$, $dA=xdy$ for $I_{x}$, while $dA=ydx$ for $I_{y}$

### Use the integration rule to derive it analytically 

For $I_x$, $dA=xdy=bdy$, as the width is constant if the small element is moving along the y axis 

$I_{x}=\iint \limits _{R}y^{2}\,\mathrm {d} A=\int _{-{\frac {b}{2}}}^{\frac {b}{2}}(\int _{-{\frac {h}{2}}}^{\frac {h}{2}}y^{2}\,\mathrm {d} y)\,\mathrm {d} x={\frac {bh^{3}}{12}} $

###  Use numerical intergration function to calculate it numerically

For example, after simplification, we can use integration program to do it here for 

$I_{x}=\iint \limits _{R}y^{2}\,\mathrm {d} A=\int _{-{\frac {b}{2}}}^{\frac {b}{2}}(\int _{-{\frac {h}{2}}}^{\frac {h}{2}}y^{2}\,\mathrm {d} y)\,\mathrm {d} x=b(\int _{-{\frac {h}{2}}}^{\frac {h}{2}}y^{2}\,\mathrm {d} y)\,\mathrm {d} x $
